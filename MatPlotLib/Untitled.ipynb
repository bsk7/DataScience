```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
#plotting using matplotlib

#plt scatter
plt.scatter(x,y,c='g')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Graph in 2D')
plt.savefig('test.png')
```

```python
y=x*x
#plt plot
plt.plot(x,y,'r')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Graph in 2D')
plt.savefig('Plottest.png')
```

```python
##creating subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'r-*')
plt.subplot(2,2,4)
plt.plot(x,y,'r--')
plt.subplot(2,2,3)
plt.plot(x,y,'go')
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sine Wave form')
plt.plot(x,y)
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.subplot(2,1,2)
plt.plot(x,y_cos,'r-*')
```

```python
#Bar plot
x=[2,4,6,8]
y=[19,23,22,33]
plt.bar(x,y)
x2=[3,5,7,9]
y2=[22,43,21,12]
plt.bar(x2,y2,color='g')
```

```python
#Histogram
a=np.array([22,34,56,67,78,44,22,33,21,64,32])
plt.hist(a)
```

```python
#BOX PLOT
data = [np.random.normal(0,std,100)for std in range(1,4)]

plt.boxplot(data,vert=True,patch_artist=True)
```

```python
#Pie chart
# data to plot
labels='python','c++','java','ruby'
sizes=[213,130,200,100]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
#explode=[0.1,0,0,0]

#plot
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
